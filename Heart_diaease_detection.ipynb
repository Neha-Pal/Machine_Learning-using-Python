```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection


```python
data=pd.read_csv('/content/heart_disease_detection.csv')
```

1--> defective ❤


0--> healthy ❤

```python
x= data.drop(columns='target',axis=1)
y=data['target']
```

```python
print(x)
```

```python
print(y)
```

splliting

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify = y,random_state = 2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model training

# Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

Model evaluation

Accuracy Score

```python
#accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy score: on Training Data:",training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy score: on Test Data:",test_data_accuracy)
```

Building a predictive system

```python
input_data=  (57,	1,	1,	124,	261,	0,	0,	141,	0,	0.3,	0,	0,	3	)
#change the input data to numpy array
input_data_array = np.asarray(input_data)
#reshape the array
input_reshaped = input_data_array.reshape(1,-1)
prediction=model.predict(input_reshaped)
print(prediction)
if(prediction[0] == 0):
  print('The person doesnot have any heart Disease...........He/She is healthy ❤ ')
else:
  print('The person has heart Disease...........He/She need medical treatment. ')
```
