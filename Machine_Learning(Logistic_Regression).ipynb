# Logistic Regression


```python
#data preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded=files.upload()
```

```python
data=pd.read_csv('Sample_logistic.csv')
```

```python
data.head()
```

```python
#extracting dependent and independent variable
x=data.iloc[:,[2,3]].values
y=data.iloc[:,4].values
```

```python
#splitting the dataset into training and test data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#create logistic regression object
from sklearn import linear_model,metrics
reg = linear_model.LogisticRegression()
```

```python
# train the model using the training sets
reg.fit(x_train, y_train)
```

```python
# making predictions on the testing set
y_pred = reg.predict(x_test)
```

```python
# compare actual response values (y_test)
# with predicted response values (y_pred)
print("Logistic Regression model accuracy(in %):",
      metrics.accuracy_score(y_test, y_pred)*100)
```
