```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#importing dataset
data = pd.read_csv('/content/creditcard.csv')
```

```python
data.head(5)
```

```python
data.tail(5)
```

```python
#dataset info
data.info()
```

```python
data.describe()
```

```python
#checking missing values
data.isnull().sum()
```

```python
#visualize the transaction
data['Class'].value_counts()
```

highly imbalanced dataset

0 --> normal transaction

1 --> fraud transaction

```python
#separate the data for analysis
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaction
data.groupby('Class').mean()
```

```python
#under-sampling
#build a sample dataset gor normal and fraudulent transactions
legit_sample = legit.sample(n = 38)
```

```python
new_data = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

splitting thw data

```python
X = new_data.drop(columns = 'Class',axis = 1)
Y = new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into train and test data


```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state = 2,test_size=.2,stratify = Y)
```

```python
print(x_train.shape,X.shape,x_test.shape)
```

Logistic Regression

```python
#training
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
X_train_pred = model.predict(x_train)
training_accuracy = accuracy_score(X_train_pred,y_train)
```

```python
print("Accuracy on training data: ",training_accuracy)
```

```python
#Accuracy on test data
x_test_pred = model.predict(x_test)
test_accuracy = accuracy_score(x_test_pred,y_test)
```

```python
print("Accuracy on test data: ",test_accuracy)
```
